### Task 4.2: Circle Calculations
Ask the user for a circle's radius. Calculate and print:

- Area (π × r²)

Use 3.14159 for π


````python
# Write your code here
import math

radius = float(input("Enter the circle's radius: "))
pi = 3.14159

circumference = 2 * pi * radius
area = pi * radius ** 2

print(f"Radius: {radius}")
print(f"Circumference: {circumference:.2f}")
print(f"Area: {area:.2f}")

print(f"\nUsing math.pi:")
print(f"Circumference: {2 * math.pi * radius:.2f}")
print(f"Area: {math.pi * radius ** 2:.2f}")
````

### Task 10.2: Phone Book
Create a dictionary that acts as a phone book with at least 4 contacts.

Then:
1. Add a new contact
2. Delete one contact using pop()
3. Use get() to safely search for a contact that might not exist
4. Print all names (keys) in the phone book


````python
# Write your code here

phone_book = {
    "John": "555-1234",
    "Alice": "555-5678",
    "Bob": "555-9012",
    "Emma": "555-3456"
}

print(f"Original phone book: {phone_book}")

phone_book["Charlie"] = "555-7890"
print(f"After adding Charlie: {phone_book}")

removed = phone_book.pop("Bob", "Not found")
print(f"Removed Bob: {removed}")
print(f"After removal: {phone_book}")

search_name = "Alice"
result = phone_book.get(search_name, "Contact not found")
print(f"Search for {search_name}: {result}")

search_name = "David"
result = phone_book.get(search_name, "Contact not found")
print(f"Search for {search_name}: {result}")

print(f"All contacts: {list(phone_book.keys())}")
````

### Challenge 1: Grade Calculator
Create a program that:
1. Asks the user for 5 test scores (use input and convert to integers)
2. Stores them in a list
3. Calculates and prints:
   - Average score
   - Highest score
   - Lowest score
   - Number of scores above 70


````python
# Write your code here
scores = []

print("Enter 5 test scores:")
for i in range(5):
    score = int(input(f"Score {i+1}: "))
    scores.append(score)

average = sum(scores) / len(scores)
highest = max(scores)
lowest = min(scores)
above_70 = sum(1 for score in scores if score > 70)

print(f"\nResults:")
print(f"Scores: {scores}")
print(f"Average: {average:.1f}")
print(f"Highest: {highest}")
print(f"Lowest: {lowest}")
print(f"Number of scores above 70: {above_70}")
````

### Challenge 2: Shopping Cart
Create a shopping cart system using a dictionary where:
- Keys are item names
- Values are dictionaries containing "price" and "quantity"

Example:
```python
cart = {
    "Apple": {"price": 2.5, "quantity": 3},
    "Bread": {"price": 3.0, "quantity": 1}
}
```

Calculate and print:
1. Total cost for each item (price × quantity)
2. Total cost of the entire cart
3. Number of items in the cart


````python
# Write your code here
# Shopping cart system
cart = {
    "Apple": {"price": 2.5, "quantity": 3},
    "Bread": {"price": 3.0, "quantity": 1}
}

print("Shopping Cart:")
print("-" * 40)
print(f"{'Item':<15} {'Price':<10} {'Qty':<5} {'Total':<10}")
print("-" * 40)

total_cost = 0
total_items = 0

for item, details in cart.items():
    item_total = details["price"] * details["quantity"]
    total_cost += item_total
    total_items += details["quantity"]

    print(f"{item:<15} ${details['price']:<9.2f} {details['quantity']:<5} ${item_total:<9.2f}")

print("-" * 40)
print(f"{'TOTAL':<15} {'':<15} ${total_cost:<9.2f}")
print(f"Total items in cart: {total_items}")
````

### Challenge 3: Word Counter
Ask the user to enter a sentence. Then:
1. Split the sentence into words (use `.split()`)
2. Count how many words there are
3. Find the longest word
4. Create a dictionary showing how many times each word appears

**Hint:** Convert everything to lowercase for accurate counting


````python
# Write your code here
# Word counter
sentence = input("Enter a sentence: ")

words = sentence.lower().split()

word_count = len(words)
print(f"\nTotal words: {word_count}")

longest_word = max(words, key=len)
print(f"Longest word: '{longest_word}' ({len(longest_word)} letters)")

word_freq = {}
for word in words:
    word = word.strip('.,!?;:')
    word_freq[word] = word_freq.get(word, 0) + 1

print("\nWord frequencies:")
for word, count in sorted(word_freq.items(), key=lambda x: x[1], reverse=True):
    print(f"  '{word}': {count} time(s)")
````
