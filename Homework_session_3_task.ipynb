### Exercise A1: While Loops (10 points)

Write a program that simulates a simple ATM withdrawal system:
- Start with a balance of $1000
- Keep asking the user to enter a withdrawal amount
- If the amount is greater than the balance, print "Insufficient funds"
- Otherwise, subtract the amount from the balance and show the new balance
- Stop when the user enters 0

**Expected interaction:**
```
Current balance: $1000
Enter withdrawal amount (0 to quit): 200
Withdrawal successful. New balance: $800

Current balance: $800
Enter withdrawal amount (0 to quit): 900
Insufficient funds!
...
```

````python
# Your code here
balance = 1000

while True:
    print(f"Current balance: ${balance}")
    try:
        amount = float(input("Enter withdrawal amount (0 to quit): "))

        if amount == 0:
            print("Thank you for using our ATM. Goodbye!")
            break
        elif amount < 0:
            print("Please enter a positive amount.")
        elif amount > balance:
            print("Insufficient funds!")
        else:
            balance -= amount
            print(f"Withdrawal successful. New balance: ${balance}")
    except ValueError:
        print("Please enter a valid number.")
````

**Hint:** A perfect square is a number whose square root is an integer (e.g., 4, 9, 16, 25)

Create a program that prints a diamond pattern with asterisks.

**Expected output for n=5:**
```
    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
```

````python
# Your code here
n = 5

for i in range(1, n + 1):
    for j in range(n - i):
        print(" ", end="")
    for j in range(2 * i - 1):
        print("*", end="")
    print()

for i in range(n - 1, 0, -1):
    for j in range(n - i):
        print(" ", end="")
    for j in range(2 * i - 1):
        print("*", end="")
    print()
````

### Exercise B2: Functions with Default Parameters (8 points)

Create a function called `calculate_shipping()` that:
- Takes three parameters: `weight`, `distance`, and `express=False`
- Base shipping cost is: `weight * 0.5 + distance * 0.1`
- If express is True, add 50% to the total cost
- Returns the total shipping cost rounded to 2 decimal places

Test with:
- `calculate_shipping(10, 100)` → should return base cost
- `calculate_shipping(10, 100, express=True)` → should return base cost × 1.5

````python
# Your code here
def calculate_shipping(weight, distance, express=False):

    base_cost = weight * 0.5 + distance * 0.1

    if express:
        total_cost = base_cost * 1.5
    else:
        total_cost = base_cost

    return round(total_cost, 2)

print(f"Standard shipping: ${calculate_shipping(10, 100)}")
print(f"Express shipping: ${calculate_shipping(10, 100, express=True)}")
````

### Exercise B3: Functions Returning Multiple Values (8 points)

Write a function called `analyze_list()` that:
- Takes a list of numbers as parameter
- Returns four values: sum, average, minimum, maximum
- Use tuple unpacking when calling the function

Test with: `numbers = [23, 45, 67, 89, 12, 34, 56, 78]`

````python
# Your code here
def analyze_list(numbers):

    if not numbers:
        return (0, 0, 0, 0)

    total = sum(numbers)
    average = total / len(numbers)
    minimum = min(numbers)
    maximum = max(numbers)

    return (total, average, minimum, maximum)

numbers = [23, 45, 67, 89, 12, 34, 56, 78]
total, avg, min_val, max_val = analyze_list(numbers)

print(f"List: {numbers}")
print(f"Sum: {total}")
print(f"Average: {avg:.2f}")
print(f"Minimum: {min_val}")
print(f"Maximum: {max_val}")
````

Given a list of student dictionaries:
```python
students = [
    {"name": "Alice", "scores": [85, 92, 78, 90]},
    {"name": "Bob", "scores": [70, 68, 72, 65]},
    {"name": "Charlie", "scores": [95, 98, 92, 96]},
    {"name": "Diana", "scores": [88, 85, 90, 87]},
]
```

Complete the following:

a) Use `map()` to create a list of student names in uppercase

b) Use `map()` to calculate the average score for each student

c) Use `filter()` to get only students with an average score above 85

d) Use `reduce()` to find the student with the highest average (you'll need `from functools import reduce`)

````python
# Setup
from functools import reduce

students = [
    {"name": "Alice", "scores": [85, 92, 78, 90]},
    {"name": "Bob", "scores": [70, 68, 72, 65]},
    {"name": "Charlie", "scores": [95, 98, 92, 96]},
    {"name": "Diana", "scores": [88, 85, 90, 87]},
]

````

````python
# Part a: Uppercase names
uppercase_names = list(map(lambda student: student["name"].upper(), students))
print("Uppercase names:", uppercase_names)
````

````python
# Part b: Average scores
averages = list(map(lambda student: {
    "name": student["name"],
    "average": sum(student["scores"]) / len(student["scores"])
}, students))
print("Average scores:", averages)
````

````python
# Part c: Students with average > 85
high_achievers = list(filter(lambda student:
    sum(student["scores"]) / len(student["scores"]) > 85, students))
print("Students with average > 85:", [s["name"] for s in high_achievers])
````

````python
# Part d: Student with highest average
highest = reduce(lambda x, y:
    x if sum(x["scores"]) / len(x["scores"]) > sum(y["scores"]) / len(y["scores"]) else y,
    students)
print(f"Student with highest average: {highest['name']} ({sum(highest['scores']) / len(highest['scores']):.2f})")
````

### Exercise B6: Practical Function - Password Validator (12 points)

Create a function called `validate_password()` that:
- Takes a password string as parameter
- Returns `True` if the password is valid, `False` otherwise
- A valid password must:
  - Be at least 8 characters long
  - Contain at least one uppercase letter
  - Contain at least one lowercase letter
  - Contain at least one digit
  - Contain at least one special character (!@#$%^&*)

Also create a function called `password_strength()` that:
- Uses `validate_password()` to check validity
- Returns "Weak" if less than 8 chars
- Returns "Medium" if valid but less than 12 chars
- Returns "Strong" if valid and 12 or more chars

Test with:
- "pass" → Weak
- "Password1!" → Medium
- "MyP@ssw0rd123" → Strong
- "lowercase" → Weak

````python
# Your code here
def validate_password(password):

    if len(password) < 8:
        return False

    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_special = any(char in "!@#$%^&*" for char in password)

    return has_upper and has_lower and has_digit and has_special

def password_strength(password):

    if len(password) < 8:
        return "Weak"

    if not validate_password(password):
        return "Weak"

    if len(password) >= 12:
        return "Strong"
    else:
        return "Medium"

test_passwords = ["pass", "Password1!", "MyP@ssw0rd123", "lowercase"]

for pwd in test_passwords:
    print(f"'{pwd}' → {password_strength(pwd)}")
    print(f"  Valid: {validate_password(pwd)}")
````

### Exercise B7: Practical Function - Shopping Cart (14 points)

Create a shopping cart system with the following functions:

1. `add_item(cart, item_name, price, quantity=1)`
   - Adds an item to the cart dictionary
   - If item exists, increase its quantity
   - Cart structure: `{"item_name": {"price": 10.99, "quantity": 2}}`

2. `remove_item(cart, item_name)`
   - Removes an item from the cart
   - Returns True if successful, False if item not found

3. `calculate_total(cart, tax_rate=0.08)`
   - Calculates subtotal, tax, and total
   - Returns a dictionary with these three values

4. `display_cart(cart)`
   - Prints the cart contents in a nice format

Test your functions by:
- Creating an empty cart
- Adding: Apple ($1.50, qty 5), Banana ($0.75, qty 3), Orange ($2.00, qty 2)
- Removing Banana
- Displaying the cart
- Calculating and displaying the total with 8% tax

````python
# Your code here
def add_item(cart, item_name, price, quantity=1):

    if item_name in cart:
        # Item exists, increase quantity
        cart[item_name]["quantity"] += quantity
    else:
        # New item
        cart[item_name] = {"price": price, "quantity": quantity}

    return cart

def remove_item(cart, item_name):

    if item_name in cart:
        del cart[item_name]
        return True
    return False

def calculate_total(cart, tax_rate=0.08):

    subtotal = sum(item["price"] * item["quantity"] for item in cart.values())
    tax = subtotal * tax_rate
    total = subtotal + tax

    return {
        "subtotal": round(subtotal, 2),
        "tax": round(tax, 2),
        "total": round(total, 2)
    }

def display_cart(cart):

    if not cart:
        print("Your cart is empty.")
        return

    print("\n" + "="*50)
    print(f"{'Item':<20} {'Price':<10} {'Qty':<5} {'Total':<10}")
    print("-"*50)

    for item_name, details in cart.items():
        item_total = details["price"] * details["quantity"]
        print(f"{item_name:<20} ${details['price']:<9.2f} {details['quantity']:<5} ${item_total:<9.2f}")

    print("="*50)

cart = {}

add_item(cart, "Apple", 1.50, 5)
add_item(cart, "Banana", 0.75, 3)
add_item(cart, "Orange", 2.00, 2)

display_cart(cart)

remove_item(cart, "Banana")
print("\nAfter removing Banana:")
display_cart(cart)

totals = calculate_total(cart)
print(f"\nSubtotal: ${totals['subtotal']}")
print(f"Tax (8%): ${totals['tax']}")
print(f"Total: ${totals['total']}")
````

## Part C: Bonus Challenge (20 points)

### Challenge: Text Analysis Tool

Create a comprehensive text analysis tool with the following functions:

1. `count_words(text)` - Returns the total number of words

2. `count_unique_words(text)` - Returns the number of unique words (case-insensitive)

3. `find_longest_word(text)` - Returns the longest word in the text

4. `word_frequency(text)` - Returns a dictionary of word frequencies (case-insensitive)

5. `most_common_words(text, n=5)` - Returns the n most common words and their counts

6. `analyze_text(text)` - Uses all the above functions to create a comprehensive report

Test your tool with this text:
```
"Python is an amazing programming language. Python is easy to learn and Python is powerful.
Many developers love Python because Python has a great community. Programming in Python is fun!"
```

Your `analyze_text()` function should produce output like:
```
=== TEXT ANALYSIS REPORT ===
Total words: ...
Unique words: ...
Longest word: ...
Most common words:
  1. python: 5
  2. is: 4
  ...
```

````python
# Your code here
import re
from collections import Counter

def count_words(text):

    words = re.findall(r'\b\w+\b', text)
    return len(words)

def count_unique_words(text):

    words = re.findall(r'\b\w+\b', text.lower())
    return len(set(words))

def find_longest_word(text):

    words = re.findall(r'\b\w+\b', text)
    if not words:
        return ""
    return max(words, key=len)

def word_frequency(text):

    words = re.findall(r'\b\w+\b', text.lower())
    return dict(Counter(words))

def most_common_words(text, n=5):

    freq = word_frequency(text)
    counter = Counter(freq)
    return counter.most_common(n)

def analyze_text(text):
    print("\n" + "="*50)
    print("=== TEXT ANALYSIS REPORT ===")
    print("="*50)

    total = count_words(text)
    unique = count_unique_words(text)
    longest = find_longest_word(text)
    common = most_common_words(text, 5)

    print(f"Total words: {total}")
    print(f"Unique words: {unique}")
    print(f"Longest word: '{longest}'")
    print("\nMost common words:")

    for i, (word, count) in enumerate(common, 1):
        print(f"  {i}. {word}: {count}")

    print("="*50)

sample_text = """Python is an amazing programming language. Python is easy to learn and Python is powerful.
Many developers love Python because Python has a great community. Programming in Python is fun!"""

analyze_text(sample_text)

print("\n" + "="*50)
print("=== WORD FREQUENCY DETAILS ===")
freq = word_frequency(sample_text)
for word, count in sorted(freq.items(), key=lambda x: x[1], reverse=True)[:10]:
    print(f"  {word}: {count}")
````
